## [Distributions](http://stanford.edu/~mwaskom/software/seaborn/tutorial/distributions.html)


```python
sns.distplot(x, kde=False, fit=stats.weibull_min);  #bins, hist, kde, rug, fit
plt.gcf().set_size_inches(3,3)
```

## [Linear relations](http://stanford.edu/~mwaskom/software/seaborn/tutorial/regression.html)  

**`regplot()  lmplot()`**

```python
sns.regplot(x='x', y='y', data=df, order=2);
plt.gcf().set_size_inches(3,3)
```
